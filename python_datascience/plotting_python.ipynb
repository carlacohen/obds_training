#### Matplotlib

- Basic plotting package 
- Very customisable - every single line/pixel 
- Need alot of code to write a simple plot

```python
# Make x & y variables using numpy 
x = np.arange(0, 10, 0.1)
y = np.sin(x)

print('This is x:')
print(x)

print()

print('This is y:')
print(y)
```

### Matplot lib object orientated interface

```python

fig = plt.figure()   # create figure object - contain axes, graphics, text
ax = plt.axes()      # create axes - a single pane

ax.plot(x, y) # create continuous line plot

ax.set_xlim(0,10)
ax.set_ylim(-1.5,1.5)
ax.set_title('title of plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.show()
```

### Matplot lib Pyplot interface

```python
plt.plot(x, y)
plt.xlim(0,10)
plt.ylim(-1.5,1.5)
plt.title('title of plt')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.show()
```

#### Scatterplot

```python
tips = sns.load_dataset("tips")
display(tips)

# set color to Yes/No depending on value in 'smoker column' 
sns.relplot(x="total_bill", 
            y="tip", 
            hue="smoker", 
            data=tips)

# as above plus shape of points for time column
sns.relplot(x="total_bill",
            y="tip",
            hue="smoker",
            style="time",
            data=tips);
```

```python

# col = Time -> create 2 plots (facets) split by time column 
sns.relplot(x="total_bill", 
            y="tip", 
            hue="smoker", 
            col="time", 
            data=tips)
```

#### Pairs plot

```python
iris = sns.load_dataset("iris")
display(iris.head())

# plot pairs plot
# assign plot to object 'g' - sets up grid
g = sns.PairGrid(iris, hue="species")

# plot histogram on diagonal
g.map_diag(plt.hist) 
# plot scatter plots in the offdiagonals
g.map_offdiag(plt.scatter)
# add legend to plot
g.add_legend();
```

### Lineplot example

```python
fmri = sns.load_dataset("fmri")
display(fmri.head())
sns.relplot(x="timepoint",
            y="signal", 
            hue="event", 
            kind="line", 
            data=fmri)

sns.relplot(x="timepoint", 
            y="signal", 
            hue="region", 
            style="event",
            dashes=False, 
            markers=True, 
            kind="line", 
            data=fmri);
```

### Facets

```python
sns.relplot(x="timepoint",
            y="signal",
            hue="subject",
            col="region",
            row="event",
            height=3,
            kind="line",
            estimator=None,
            data=fmri)
```

### Plot Categorical Data

```python
# create violin plots
g = sns.catplot(x="day", 
                y="total_bill", 
                kind="violin",
                inner=None,
                data=tips)

# add individual datapoints -> provide ax argument
sns.swarmplot(x="day",
              y="total_bill",
              color="k",
              size=3,
              data=tips,
              ax=g.ax) 
```

Rename columns and remove unwanted columns (Chr, Start, End, Strand, Length) 
#This is because we want to see a table that has each gene and then the number of counts for each sample (bam file)


```python
df.columns #generate a list of columns
```

```python
counts = df.drop(columns=["Chr", "Start", "End", "Strand", "Length"]) #remove unwanted columns

#David's method, first use string replace to remove the sam and bam bits (actually superfluous because of what follows, but useful to know!)
counts.columns = counts.columns.str.replace("sam/", "").str.replace(".bam", "") #replace "sam" with nothing, replace "bam" with nothing
#Charlie's method to rename columns
#using {} means it is a dictionary
#inplace=True means it edits the table
counts.rename(columns={"ERR1755082": "cd4_r1", "ERR1755083": "cd4_r2", "ERR1755084": "cd4_r3", "ERR1755085": "cd8_r1", "ERR1755086": "cd8_r2", "ERR1755087": "cd8_r3"}, inplace=True)
counts
```

Filter out genes with no counts


```python
#Set GeneID as the index
counts = counts.set_index("Geneid")
```

```python
#Sum columns 2-7 for each row
#don't need to specify which columns as geneid is not an integer anyway
counts["gene_count"]=counts.sum(axis=1)
counts
```

```python
#Then select instances where that column > 0 and set as a new variable
#also remove the column gene_count
filtered_counts = counts[counts["gene_count"]>0].drop(columns=["gene_count"])
filtered_counts
```

Plot scatter plot of correlation of replicates (pairs plot) (you will need to google how to show plots in jupyter notebook)
- %matplotlib inline (done earlier)
- This allows us to compare the 3 replicates for each cell type one by one and across cell types
- It shows that the replicates within cell types are more closely related than those across cell types, as we would expect
- Histograms on the diagonal are there because you cannot meaningfully correlate a sample with itself
- The histogram is showing the distribution of gene counts



```python
g = sns.PairGrid(np.log2(filtered_counts +1))

# plot histogram on diagonal
#made the bins smaller (from no comment to 100) but they are still not v informative
#So plot the log instead (in line 1 using np.log2) 
#We had do to +1 because you can't plot a log of 0
#(alternatively we could have increased our filtering threshold)
g.map_diag(plt.hist, bins=100) 
# plot scatter plots in the offdiagonals
g.map_offdiag(plt.scatter)
# add legend to plot
g.add_legend()
```

Normalise read counts to total reads per sample & convert to 'counts per million (CPM)'


```python
sum_counts = filtered_counts_gene_index.sum() #sum the columns i.e. total counts for each sample
display(filtered_counts_gene_index)
display(sum_counts)
filtered_counts_CPM = filtered_counts_gene_index.div(sum_counts, axis=1)*1e6 #divide the df by the sum of the columns & multiply by 1000000
filtered_counts_CPM
```

Log transform normalised read counts


```python
log2_filtered_counts_CPM = np.log2(filtered_counts_CPM +1)
log2_filtered_counts_CPM
```

Create new dataframe of 100 genes with highest average variance in log cpm across conditions
- these might be the most interesting ones to look at


```python
log2_filtered_counts_CPM["variance"] = log2_filtered_counts_CPM.var(axis=1)
log2_filtered_counts_CPM
top_100_variance = log2_filtered_counts_CPM.sort_values(by=["variance"], ascending=False).head(100)
top_100_variance
```

Plot the of expression of the top 5 most variable genes (log cpm) across all samples (line plot)


```python
top_5_variance_melt = top_100_variance.head(5) \
                                     .drop("variance", axis=1) \
                                     .reset_index() \
                                     .melt(id_vars=["Geneid"], var_name="sample", value_name="counts")
top_5_variance_melt

sns.lineplot(data=top_5_variance_melt, x="Geneid", y="counts", hue="sample")
plt.show()
sns.lineplot(data=top_5_variance_melt, x="sample", y="counts", hue="Geneid")
```
